### Importing packages and data

```python
from gensim.parsing.preprocessing import remove_stopwords, preprocess_string
import pandas as pd
import numpy as np
import os
```

```python
twitter=pd.read_csv(r"C:\data location")
twitter['comment'] = twitter['content']
twitter.head()
```

### Filter by English

```python
twitter = twitter[twitter['language'] == 'en']
twitter.head()
```

### Remove stopwords

```python
twitter['comment'] = twitter['comment'].apply(lambda x: remove_stopwords(x))

twitter.head()
```

### Make lowercase

```python
twitter['comment'] = twitter['comment'].apply(lambda x: x.lower())

twitter.head()
```

### Removing URLs

```python
import re
def removing_URLs(d):
    return re.sub('((www.[^s]+)|(https?://[^s]+))',' ', d)
```

```python
twitter['comment'] = twitter['comment'].apply(lambda x: removing_URLs(x))

twitter.head()
```

### Removing Punctuation

```python
import string
def removing_punctuations(text):
    translator = str.maketrans('', '', string.punctuation)
    return text.translate(translator)
```

```python
twitter['comment'] = twitter['comment'].apply(lambda x: removing_punctuations(x))

twitter.head()
```

### Removing Numbers

```python
def removing_numbers(data):
    return re.sub('[0-9]+', '', data)
```

```python
twitter['comment'] = twitter['comment'].apply(lambda x: removing_numbers(x))

twitter.head()
```

### Removing Apostrophe

```python
def removing_apos(data):
    return re.sub('[’]+', '', data)
```

```python
twitter['comment'] = twitter['comment'].apply(lambda x: removing_apos(x))

twitter.head()
```

### Truncate Comments over 500 characters

```python
twitter['comment'] = twitter['comment'].apply(lambda x: x[:500])
twitter.head()
```

### Remove Emojis

```python
emoji_pattern = re.compile("["
        u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map symbols
        u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
        u"\U0001F1F2-\U0001F1F4"  # Macau flag
        u"\U0001F1E6-\U0001F1FF"  # flags
        u"\U0001F600-\U0001F64F"
        u"\U00002702-\U000027B0"
        u"\U000024C2-\U0001F251"
        u"\U0001f926-\U0001f937"
        u"\U0001F1F2"
        u"\U0001F1F4"
        u"\U0001F620"
        u"\u200d"
        u"\u2640-\u2642"
        "]+", flags=re.UNICODE)

twitter['comment'] = twitter['comment'].apply(lambda x: emoji_pattern.sub(r'', x))
twitter.head()
```

### Removing @handles

```python
def remove_handles(data):
    return re.sub('@[^\s]+','',data)
```

```python
twitter['content'] = twitter['content'].apply(lambda x: remove_handles(x))

twitter.head()
```

```python
os.chdir(r'C:\file location')
twitter.to_csv('twitter_clean.csv')
```
